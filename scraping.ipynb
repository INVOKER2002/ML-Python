```python
import pandas as pd 
from bs4 import BeautifulSoup
```

```python
f = open('home.html','r')
```

```python
soup = BeautifulSoup(f , 'lxml')
```

```python
tags = soup.find_all('h5')
print(tags)
#finds all the h5 tags in the webpage htmml file 
#to find the first tag use just the find method()
```

```python
course_cards = soup.find_all('div', class_='card') #filtering the div tags whicch containg the class cards 
```

```python
for course in course_cards:
    print(course.text)
    course_name = course.h5.txt
    print(course_name)
    course_price= course.a.txt
    print(course_price)
```
